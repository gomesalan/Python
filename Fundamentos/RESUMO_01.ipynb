#### 4.2) Ordem de precedência

```python
print(2 + 3 * 4 - 5) # 2 + (3.4) - 5

# O uso de parênteses altera essa ordem 

print(2 + 3 * (4 - 5))
```

```python
# Ordem de associatividade - esquerda para direita (regra geral). 

print(2 - 3 + 4)
print(4 + 3 - 2)
```

```python
# O mesmo vale para divisão - esquerda para direita

print(12/2/3/4) # 12/2 = 6, 6/3 = 2 , 2/4 = 0.5 
print((12/2)/(3/4)) # 6/0.75 = 8.0 
```

```python
# Ordem de associatividade - direita para esquerda (exceção é a potenciação)

print(2**2**3) # primeiro 2**3 = 8 e depois 2**8 = 256
```

```python
# Em relação a divisão inteira ou resto da divisão, prioridade secundária. 

12 * 2 % 10 # 12*2 = 24 e 24 % 10 = 4 
```

#### 10.2) Manipulação de strings através do seu índice

```python
print(string1) 
print(string1[0]) # Primeiro elemento da string (em python a indexação começa por 0 )
print(string1[1]) # Segundo elemento da string
```

```python
# Indexação negativa - 'trás para frente' 

print(string1) 
print(string1[-1]) # Último elemento da string
print(string1[-2]) # Penúltimo elemento da string
```

```python
# Imutabilidade da string:

string1[0] = 'J' # (TypeError: 'str' object does not support item assignment)
```

#### 10.3) Fatiamento de strings - slicing 

```python
# string[a:b] - [a,b)

print(string1)
print(string1[0:3]) # [0,3)
print(string1[1:6]) # [1,6)
```

```python
# Ao omitir a e/ou b, são considerados todos elementos desde o ínicio e/ou final.

print(string1[:6]) # todos elementos até a posição 6 (índice 5)  
print(string1[6:]) # todos elementos a partir da posição 6 (índice 6)
```

```python
# Retornar tudo, exceto a última letra 

print(string1[:-1])
print(string1[:-2])
print(string1[-6:]) # índice -6 até o final 
print(string1[11:]) # índice 11 até o final
```

#### 10.4) Fatiamento com passo em strings

```python
# Todos elementos com passo descrecente 
print(string2[::-1])
print(string2[::-2])
```

#### 10.5) Métodos built-in para strings

Python é uma linguagem orientada a objeto, portanto, as strings como objetos da linguagem, possuem atributos e métodos. 

```python
string = "Universidade Federal do ABC"
```

```python
# Dividir a string - método split() - sem argumento o padrão é por espaços em branco. 
# Gera uma lista com os elementos da string. 

string.split()
```

```python
# Observe que 'a' não está na lista, pois foi usado como critério. 
string.split('a')
```

```python
string.split('A')
```

```python
# Upper - letras maiúsculas 
string.upper()
```

```python
# Lower - letras minúsculas
string.lower()
```

```python
# Contar número de ocorrências de determinado caractere - método count() 
string.count('a') 
```

```python
# Encontra o índice de um elemento especificado.
string.find("v") 
```

```python
# Centraliza a string - center(N de elementos da string final, 'padrão')
string.center(50, '*')
```

```python
# Note que "Universidade Federal do ABC" tem 27 elementos. 
len(string)
```

```python
# Método replace() - replace ('a ser substituído','seu substituto')

tel = "(11)94665-2219"
tel_semhifen = tel.replace('-','')
print(tel_semhifen) 

# podemos aplicar simultaneamente.
tel = tel_semhifen.replace('(','').replace(')','')  
print(tel) 
```

```python
# Método join() - utiliza uma lista. 

'/'.join(['02', '10', '2021'])
```

```python
# Pode ser alterado utilizando métodos:

conj1.remove(1)
```

```python
conj1
```

```python
# Acresentando elementos 

conj1.update({99})
conj1
```
