## Funções especiais 

### 1 - Função lambda

Permitem criar funções anônimas, ou seja, que não precisam ser definidas. 

lambda (parâmetro de entrada) : (retorno)

```python
# Exemplo 
quadrado = lambda x: x**2 
quadrado(3)
```

```python
# Exemplo
soma = lambda x,y: x + y 
soma(2,2)
```

```python
# Exemplo
maior = lambda x, y : x if (x > y) else y
maior(3,5)
```

```python
# Exemplo
par = lambda n: 1 if (n%2 == 0) else 0
valor = 12
if(par(valor)):
    print("PAR")
else:
    print("IMPAR")
```

```python
# Exemplo
inverte_string = lambda s : s[::-1]
texto = "Este é um texto em python"
print(inverte_string(texto))
```

### 2 - Map()

Aplica uma função a cada elemento de uma sequência e retorna uma lista com os resultados. 

```python
# Exemplo

def celsius_to_fahrenheit(t):
    return round((5/9)*(t - 32), 1)
 
temp = [20, 40, 67, 100, 180, 500, 13]

map(celsius_to_fahrenheit, temp)  # retorna um iterator  
```

```python
# Para obter o resultado é necessário usar list().

list(map(celsius_to_fahrenheit, temp))
```

#### 2.1) map() com expressão lambda

```python
temp = [20, 40, 67, 100, 180, 500, 13]

list(map(lambda t: (5/9)*(t - 32), temp))
```

```python
valor_venda = [15, 19, 21, 73]
valor_compra = [10, 12, 13, 40]

lucro = list(map(lambda x, y : x - y, valor_venda, valor_compra))
lucro
```

### 3 - Reduce()

Aplica uma função a cada elemento de uma sequência, até que reste apenas um elemento.

```python
# Deve ser importada do pacote functools:
from functools import reduce 
```

```python
# Exemplo:

lista = [11, 12, 13, 14, 15]

def soma(a, b):
    return a + b 

reduce(soma, lista) 

# O que foi feito:
# 1°) 11 + 12 = 23
# 2°) 23 + 13 = 36 
# 3°) 36 + 14 = 50
# 4°) 50 + 15 = 65 
```

```python
minha_lista = [2, 4, 5, 2, 1]

def achar_minimo(x, y):
    if x < y:
        return x
    else:
        return y

reduce(achar_minimo, minha_lista)
```

#### 3.1) reduce() com expressão lambda

```python
# Verificando qual o maior valor em uma lista.

alturas = [1.60, 1.78, 1.80, 1.96, 1.77, 1.85]

reduce( lambda x, y : x if (x > y) else y , alturas)
```

###  4 - Filter()

Filtra todos elementos de uma sequência para os quais a função retorne verdadeiro.  

```python
# Exemplo: verificar os elementos ímpares de uma lista.

lista_num = [2, 3, 4, 5, 77, 63, 121, 30, 48]

def impar(n):
    if n % 2 != 0:
        return True 
    else:
        return False

filter(impar, lista_num) # Retorna um iterator. 
```

```python
# Usar junto com a função list().
list(filter(impar, lista_num))
```

#### 3.1) Filter() com expressão lambda

```python
# Retornar apenas os elementos pares. 

list(filter( lambda n: n%2 == 0, lista_num))
```

### 4 - List Comprehension 

Aplica um expressão arbitrária a uma sequência de elementos. Essencialmente é um loop for() contruído dentro de colchetes []

```python
# Exemplo: criar uma lista com o quadrado dos 10 primeiros naturais.

[x**2 for x in range(11)]
```

```python
# Exemplo 

[x**2 for x in range(11) if x > 3]
```

```python
# Exemplo: criar uma lista com os quadrados pares dos 10 primeiros naturais. 

[x**2 for x in range(11) if x%2 == 0]
```

#### 4.1) Operações aninhadas 

```python
# Exemplo 

['menor' if x < 50 else "maior" for x in [x**2 for x in range(11)]]
```

```python
# 1° - entre parênteses é calculada uma lista com o quadrado dos 10 primeiro números naturais 

lista_quarta = [x**2 for x in (i**2 for i in range(11))]
lista_quarta 
```

```python
# Retorna uma lista com cada caracter de uma string.

list_string = [j for j in "Esta é uma string"]
list_string
```

### 5 -  Zip()

Une duas sequências e retorna uma lista de tuplas com os pares de elementos de cada seqûencia.

```python
a = [1, 2, 3]
b = [4, 5, 6]

zip(a,b) 
```

```python
# Usando a função list().
list(zip(a,b))
```

```python
# Sequências com número diferente de elementos.

a = [1, 2, 3]
b = [4, 5, 6, 7, 8]

list(zip(a,b))
```

```python
# Usando com dicionários.

d1 = {'A': 1,'B': 2}
d2 = {'C': 3,'D': 4}
```

```python
list(zip(d1.keys(), d2.keys()))
```

```python
list(zip(d1.values(), d2.values()))
```

```python
list(zip(d1, d2.values()))
```

```python
# Dicionário d1 passa a ter os valores do diciário d2. 

def troca(d1, d2):
    d3 = {}
    for k, v in zip(d1, d2.values()):
        d3[k] = v
    return d3
    
troca(d1, d2)
```

### 6 - Enumerate()

Recebe uma sequência e retorna uma tupla com o par índice/elemento da sequência. 

```python
lista_exemplo = [1, 2, 3, 4, 5]
```

```python
# Retorna um iterator.

enumerate(lista_exemplo)
```

```python
# Usando a função list()

list(enumerate(lista_exemplo))
```

```python
# Acessando a lista de tuplas. 

l = list(enumerate(lista_exemplo))
print(l[0])
print(l[0][0])
```

```python
lista_letra = ['a', 'b', 'c', 'd']

list(enumerate(lista_letra))
```
